```python
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk import FreqDist

nltk.download('gutenberg')
nltk.download('punkt')

text = gutenberg.raw('austen-emma.txt')
tokens = word_tokenize(text)

wh_words = [word.lower() for word in tokens if word.lower().startswith('wh') and len(word) > 2]
print("WH-Words:", wh_words)
```

```python
unique_wh_words = list(set(wh_words))
print("Uique-WH-Words:", unique_wh_words)
```

```python
duplicates = [word for word in wh_words if text.lower().count(word) != tokens.count(word.lower())]
print("Duplicates:", duplicates)
```
